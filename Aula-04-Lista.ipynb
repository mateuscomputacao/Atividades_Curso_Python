### Exercício 1: Verificação Simples

Crie uma função em Python que receba um número inteiro como entrada e retorne `True` se o número for positivo, e `False` caso contrário. Considere 0 como não positivo.

```python
numero = int(input("Digite um número inteiro: "))

if numero == 0:
  booleano = False
elif numero % 2 == 0:
  booleano = True
else:
  booleano = False

print(booleano)
```

### Exercício 3: Ano Bissexto

Crie uma função que determine se um ano fornecido é um ano bissexto. Um ano é bissexto se for divisível por 4, mas não por 100, a menos que também seja divisível por 400.

```python
ano = int(input("Digite um ano: "))

if (ano % 4 == 0 and ano % 100 != 0) or ano % 400 == 0:
  print("\nO ano é bissexto.")
else:
  print("\nO ano não é bissexto.")
```

### Exercício 5: Ponto Dentro de Retângulo

Escreva uma função que receba as coordenadas de um ponto `(x, y)` e as coordenadas de dois pontos que definem um retângulo `(x1, y1)` e `(x2, y2)`, onde `(x1, y1)` é o canto inferior esquerdo e `(x2, y2)` é o canto superior direito. A função deve retornar `True` se o ponto `(x, y)` estiver **dentro ou na borda** do retângulo, e `False` caso contrário.

```python
x_origem_1 = int(input("Digite a cordenada -X1- que define o canto infereior/esquerdo do retângulo: "))
y_origem_1 = int(input("Digite a cordenada -Y1- que define o canto infereior/esquerdo do retângulo: "))
x_origem_2 = int(input("Digite a cordenada -X2- que define o canto superior/direito do retângulo: "))
y_origem_2 = int(input("Digite a cordenada -Y2- que define o canto superior/direito do retângulo: "))
x_ponto = int(input("\nDigite a cordenada -X- do ponto a ser descoberto: "))
y_ponto = int(input("Digite a cordenada -Y- do ponto a ser descoberto: "))

if x_origem_1 <= x_ponto <= x_origem_2 and y_origem_1 <= y_ponto <= y_origem_2:
  print("\nO ponto está dentro ou na borda do retângulo.")
else:
  print("\nO ponto está fora do retângulo.")
```

### Exercício 6: Tipo de Triângulo com Validação

Desenvolva uma função que receba três números representando os comprimentos dos lados de um triângulo. A função deve primeiro verificar se os lados podem formar um triângulo (a soma de quaisquer dois lados deve ser maior que o terceiro lado). Se for um triângulo válido, ela deve retornar:
- 'Equilátero' (todos os lados iguais)
- 'Isósceles' (dois lados iguais)
- 'Escaleno' (todos os lados diferentes)
- 'Não é um triângulo válido' (se não atender à condição de formação do triângulo).

Combine as condições usando operadores lógicos.

```python
lado_a = int(input("Digite o tamanho do primeiro lado: "))
lado_b = int(input("Digite o tamanho do segundo lado: "))
lado_c = int(input("Digite o tamanho do terceiro lado: "))

a = lado_a + lado_b
b = lado_b + lado_c
c = lado_a + lado_c

if a > lado_c and b > lado_a and c > lado_b:
  print("As medidas formam um triângulo!")
  if lado_a == lado_b and lado_b == lado_c and lado_c == lado_a:
    print("O triângulo é equilátero.")

  else:
    if lado_a != lado_b and lado_b != lado_c and lado_c != lado_a:
      print("O triângulo é escaleno.]")

    else:
      print("O triângulo é isósceles.")
else:
  print("As medidas não formam um triângulo")
```

### Exercício 7: Elegibilidade para Desconto de Idosos

Crie uma função que receba a idade de uma pessoa (inteiro) e um valor booleano indicando se ela é membro de um programa de fidelidade (`is_member`). A função deve retornar `True` se a pessoa for elegível para um desconto, e `False` caso contrário. Uma pessoa é elegível se:
- Tiver 65 anos ou mais **E** for membro do programa.
- OU tiver 70 anos ou mais (independentemente de ser membro).

Combine as condições usando operadores lógicos.

```python
def verificar_desconto_idoso(idade, membro):
    condicao1 = (idade >= 65 and membro)
    condicao2 = (idade >= 70)
    return condicao1 or condicao2


idade = int(input("Digite a idade: "))
membro = input("É membro? (s/n): ").lower() == 's'
print(f"Elegível para desconto (idade={idade}, membro={membro}): {verificar_desconto_idoso(idade, membro)}")

```

### Exercício 8: Cálculo de Nota com Condições Múltiplas

Implemente uma função que receba uma pontuação (um número inteiro entre 0 e 100) e retorne a letra correspondente à nota, seguindo estas regras, e também valide a pontuação:
- A: 90-100
- B: 80-89
- C: 70-79
- D: 60-69
- E: Abaixo de 60
- 'Pontuação Inválida': Se a pontuação estiver fora do intervalo de 0 a 100.

Combine as verificações de intervalo com operadores lógicos.

```python
pontuacao = int(input("Digite sua pontuação (de 0 a 100): "))

if pontuacao < 0 or pontuacao > 100:
  print("Pontuação inválida.")
elif 90 <= pontuacao <= 100:
  print("Pontuação na classe -A-")
elif 80 <= pontuacao <= 89:
  print("Pontuação na classe -B-")
elif 70 <= pontuacao <= 79:
  print("Pontuação na classe -C-")
elif 60 <= pontuacao <= 69:
  print("Pontuação na classe -D-")
elif pontuacao < 60:
  print("Pontuação na classe -E-")
```

### Exercício 10: Aprovação de Aluno Complexa

Crie uma função em Python que receba os seguintes parâmetros:
- `nota_final` (int): A nota final do aluno (de 0 a 100).
- `frequencia_percentual` (int): A porcentagem de frequência do aluno (de 0 a 100).
- `completou_projetos` (boolean): `True` se o aluno completou todos os projetos, `False` caso contrário.

A função deve retornar `True` se o aluno for aprovado, e `False` caso contrário, seguindo as seguintes regras:
1. O aluno é aprovado se `nota_final` for maior ou igual a 70 **E** `frequencia_percentual` for maior ou igual a 75 **E** `completou_projetos` for `True`.
2. **Exceção:** Se o aluno tiver `nota_final` maior ou igual a 90, ele será aprovado **mesmo que** a `frequencia_percentual` seja menor que 75, desde que `completou_projetos` seja `True`.
3. Em todos os outros casos, o aluno é reprovado.

```python
def verificar_aprovacao():

    nota_final = int(input("Digite a nota final (0-100): "))
    frequencia_percentual = int(input("Digite a frequência (0-100): "))


    resposta_projeto = input("Completou todos os projetos? (s/n): ").lower()
    completou_projetos = resposta_projeto == 's'


    condicao_padrao = (nota_final >= 70 and frequencia_percentual >= 75)


    condicao_excecao = (nota_final >= 90)


    if (condicao_padrao or condicao_excecao) and completou_projetos:
        return True
    else:
        return False

resultado = verificar_aprovacao()
print(f"Aluno aprovado? {resultado}")
```
