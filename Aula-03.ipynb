# Aula 03: Estruturas de Decisão com Álgebra Booleana

Na aula anterior, aprendemos a fazer uma única pergunta com o `if` (ex: `idade >= 18`).
E se precisarmos que **duas coisas** sejam verdadeiras ao mesmo tempo? Ou se precisarmos que pelo menos **uma de duas** coisas seja verdade?
É aqui que entram os operadores lógicos: `and` (E), `or` (OU) e `not` (NÃO).

## 1. O operador `and` (E)

Com o `and`, a condição completa só será verdadeira se **TODAS as partes forem verdadeiras**. Pense em um sistema de login: você precisa acertar o usuário E a senha.

```python
usuario_correto = "admin"
senha_correta = "1234"

usuario = input("Usuário: ")
senha = input("Senha: ")

if usuario == usuario_correto and senha == senha_correta:
    print("Login efetuado com sucesso!")
else:
    print("Usuário ou senha incorretos.")
```

Outro exemplo clássico do `and` é verificar se um número está **entre** dois valores (um intervalo).

```python
idade = int(input("Qual sua idade? "))

# O voto no Brasil é obrigatório entre 18 (incluso) e menos de 70 anos.
if idade >= 18 and idade < 70:
    print("Seu voto é OBRIGATÓRIO.")
else:
    print("Seu voto é facultativo ou você não pode votar.")
```

*Dica Ninja Python:* O Python permite escrever a Matemática básica de um jeito mais próximo do papel para representar intervalos. A linha acima poderia ser escrita também como: `if 18 <= idade < 70:`.

## 2. O operador `or` (OU)

Com o `or`, a condição completa será verdadeira se **PELO MENOS UMA** das partes for verdadeira. Só será falso se ambas as partes estiverem falsas.

```python
dia_da_semana = input("Que dia da semana é hoje? ").lower()

if dia_da_semana == "sábado" or dia_da_semana == "domingo":
    print("Eba! É final de semana!")
else:
    print("Dia útil. Vamos trabalhar/estudar.")
```

## 3. O operador `not` (NÃO)

O `not` simplesmente **inverte** o valor lógico. Se era Verdade, vira Falso. Se era Falso, vira Verdade.
É muito útil para verificações do tipo "se NÃO está vazio" ou "se a permissão NÃO foi concedida".

```python
chovendo = False

# Se NÃO estiver chovendo...
if not chovendo:
    print("Vamos ao parque!")
else:
    print("Vamos assistir a um filme em casa.")
```

```python
nome = input("Digite seu nome: ")

# O Python permite inverter a comparação de Igualdade.
# `!=` é a mesma coisa que fazer `not (variavel == valor)`
if nome != "Visitante":
    print(f"Bem vindo de volta, {nome}!")
```
